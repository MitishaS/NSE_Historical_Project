```python
import pandas as pd
from nsepy import get_history
from datetime import date

stock= ['JINDALSTEL','CANBK','TATAMOTORS','BAJFINANCE','ADANIENT','RELIANCE','APOLLOTYRE','ESCORTS']

# As historical fututres data is displayed in3 months interval
# first we created an empty dataframe df1 for 3 months Jan2019 to Mar2019, then we will add the data to it.
df1 = pd.DataFrame()
for stock_name in stock:
    # in get_history function, Symbol means the name of stock, the start date and end date for the interval of time,
    # as we want the futures date so will take futures=True, and expiry date will denote the expiry of contract 
    
    data = get_history(symbol=stock_name, start=date(2019, 1, 1), end= date(2019,3,31), 
                       futures=True, expiry_date=date(2019,3,28))
    df1 = df1.append(data)
    
    # we will select the columns we want in output
    df1 = df1[["Symbol","Open","Close","Low","High"]]
    
    # Load the data to csv file
    df1.to_csv('case1.csv')
    print(df1)


df2 = pd.DataFrame()
for stock_name in stock:
    # similiar to previous, in this we will store the data of April2019 to June2019 in data2
    data2 = get_history(symbol=stock_name, start=date(2019, 4, 1), end= date(2019,6,30), 
                        futures=True, expiry_date=date(2019,6,27))
    # we will append the data of this for-loop with the previous dataframe
    df2 = df2.append(data2)    
    df2 = df2[["Symbol","Open","Close","Low","High"]]
    print(df2)
    # load the data to csv, mode='a' meann it will append the previous data, or else it will replace the older data stored.
    df2.to_csv('case1.csv', mode='a')



df3 = pd.DataFrame()
for stock_name in stock:
    # in this the data will be from july2019 to sept2019 
    data3 = get_history(symbol=stock_name, start=date(2019, 7, 1), end= date(2019,9,30), 
                        futures=True, expiry_date=date(2019,9,26))
    df3 = df2.append(data3) 
    df3 = df3[["Symbol","Open","Close","Low","High"]]
    print(df3)
    df3.to_csv('case1.csv', mode='a')

df4 = pd.DataFrame()
for stock_name in stock:
    data4 = get_history(symbol=stock_name, start=date(2019, 10, 1), end= date(2019,12,31), futures=True,expiry_date=date(2019,12,26))
    df4 = df3.append(data4) 
    df4 = df4[["Symbol","Open","Close","Low","High"]]
    print(df4)
    df4.to_csv('case1.csv', mode='a')

```
