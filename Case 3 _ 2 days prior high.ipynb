```python
from nsepy import get_history
from datetime import datetime, date, timedelta
import csv
import pandas as pd
import numpy as np

#extract data from case1.csv and store it in dataframe 
df = pd.read_csv('case1.csv')

#timedelta helps to substract date
d = timedelta(days=2)

#this dataframe take time from csv file and store in Date
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
Date= df['Date']

#deletes the timedelta date from the date from csv
df['pdate'] = Date - d
df['Week_Number'] = df['pdate'].dt.week

# we will sort by weeks and stocks using groupby function, 
# and will get get the store the maximum and minimun of each week in High and Low respectively.
df2 = df.groupby(['Week_Number','Symbol'],as_index=False).agg({'High':'max', 'Low':'min'})
print(df2)

#load the data in csv file
df2.to_csv('case3.csv')
```
