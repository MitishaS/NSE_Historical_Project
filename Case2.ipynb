```python
from nsepy import get_history
from datetime import datetime
import csv
import pandas as pd
import numpy as np

# extract the data from case1.csv file and store it in dataframe
df = pd.read_csv('case1.csv')

# first we will seperate the date from whole date, and sort it in weeks 
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Week_Number'] = df['Date'].dt.week

# we will sort by weeks and stocks using groupby function, 
# and will get get the store the maximum and minimun of each week in High and Low respectively.
df2 = df.groupby(['Week_Number','Symbol'],as_index=False).agg({'High':'max', 'Low':'min'})
print(df2)

#load the data in csv file
df2.to_csv('case2.csv')
```
